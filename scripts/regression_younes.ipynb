```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import des donn√©es

```python
df_fusion = pd.read_csv("../data/DVF_fusion.csv")
```

# Quelques statistiques descriptives

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogramme du nombre de d√©pendances
sns.countplot(x="nombre_dependances", data=df_fusion, ax=axes[0])
axes[0].set_title("Nombre de d√©pendances")
axes[0].set_xlabel("Nombre de d√©pendances")
axes[0].set_ylabel("Nombre de maisons")

# Histogramme du nombre de pi√®ces principales
sns.countplot(x="nombre_pieces_principales", data=df_fusion, ax=axes[1])
axes[1].set_title("Nombre de pi√®ces principales")
axes[1].set_xlabel("Nombre de pi√®ces principales")
axes[1].set_ylabel("Nombre de maisons")

plt.tight_layout()
plt.show()
```

# Cr√©ation de quelques variables

```python
df_fusion["indic_distance_littoral<10km"] = (df_fusion["distance_littoral"] <= 10000).astype(int)
df_fusion["distance_littoral<10km"] = (df_fusion["distance_littoral"] <= 10000).astype(int)*df_fusion["distance_littoral"]/1000

df_fusion["distance_fleuve_km"] = df_fusion["distance_fleuve"]/1000
df_fusion["dep_2"] = (df_fusion["nombre_dependances"] == 2).astype(int)
df_fusion["dep_3plus"] = (df_fusion["nombre_dependances"] >= 3).astype(int)
df_fusion["piece_4"] = (df_fusion["nombre_pieces_principales"] == 4).astype(int)
df_fusion["piece_5"] = (df_fusion["nombre_pieces_principales"] == 5).astype(int)
df_fusion["piece_6plus"] = (df_fusion["nombre_pieces_principales"] >= 6).astype(int)


# S√©lection des variables explicatives
variables_explicatives = [
    "distance_fleuve", "distance_littoral", "distance_mairie_km",  
    "surface_reelle_bati", "surface_terrain",
    "risque_debordement_fort", "risque_debordement_moyen", "risque_debordement_faible",
    "risque_submersion_fort", "risque_submersion_moyen", "risque_submersion_faible", "piece_2", "piece_3", "piece_4", "piece_5", "piece_6plus", "dep_1", "dep_2", "dep_3plus"]

# On enl√®ve les lignes avec des valeurs manquantes
df = df_fusion.dropna(subset=["log_prix"] + variables_explicatives)
df = df[df["nombre_pieces_principales"] >= 1]
```

# 1√®re r√©gression

```python
X = df[variables_explicatives]
y = df["log_prix"]

X_const = sm.add_constant(X)

# Ajustement du mod√®le
model = sm.OLS(y, X_const).fit()

# R√©sum√© des r√©sultats
print(model.summary())
```
